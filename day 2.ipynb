#Python Loops

```python
fruits = ["apple","banana","cherry"]
```

```python
for fruit in fruits:
  # print(fruit)
  if fruit == "banana":
    print(fruit)
    break
  else:
    print(fruit)
```

```python
for fruit in fruits:
  # print(fruit)
  if fruit == "banana":
    # print(fruit)
    continue
    print(fruit)
  else:
    print(fruit)
```

```python
i = 1
while i < 6:
  print(i)
  i += 1
  print("New value of i -->",i)
```

```python
i = 1 #3
while i < 6:
  print(i)
  i += 1
  if i == 3:
    pass

  print("New value of i -->",i)
```
